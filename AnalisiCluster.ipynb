```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Carica il dataset
df = pd.read_csv("ExportedDatasets/clusterRegioni.csv")

# Ordina il DataFrame in base a "NUMCLUSTER" in ordine decrescente
df_sorted = df.sort_values(by="NUMCLUSTER", ascending=False)

# Genera una lista di colori dalla mappa 'rainbow' di Matplotlib
num_bars = len(df_sorted)
colors = plt.cm.rainbow(np.linspace(0, 1, num_bars))

# Grafico a barre ordinato con colori arcobaleno
plt.figure(figsize=(12,6))
plt.bar(df_sorted["DEN_REG"], df_sorted["NUMCLUSTER"], color=colors)

# Etichette e titolo
plt.xlabel("Regione")
plt.ylabel("Numero di cluster", labelpad=20)
plt.title("Distribuzione dei cluster per regione")

# Ruota le etichette dell'asse X in verticale
plt.xticks(rotation=90)

# Mostra il grafico
plt.show()
```
